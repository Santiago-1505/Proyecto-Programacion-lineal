```python
import re

def parsear_expresion(texto: str) -> list[Termino]:
    patron = r'([+-]?\d*\.?\d*)x(\d+)'

    terminos = []

    for coef, var in re.findall(patron, texto.replace(" ", "")):

        if coef in ("", "+"):
            coeficiente = 1
        elif coef == "-":
            coeficiente = -1
        else:
            coeficiente = float(coef)

        variable = int(var)

        terminos.append(
            Termino(coeficiente, variable)
        )

    return terminos
```
